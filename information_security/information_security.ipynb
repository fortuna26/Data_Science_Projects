## Обучение бэйзлайн модели (Baseline model machine learning)

Сделаем минимальную, самую простую предобработку данных и обучим несколько моделей методом кросс-валидации. Получим их метрики, чтобы было от чего отталкиваться и с чем сравнивать при последующей обработке и обучении.

### Простая предобработка данных (Simple data preprocessing)

```python
# удалим строки с пропусками

data_bl = data.dropna()
```

```python
# проверим как это повлияло на целевой признак, не уменьшились ли малочисленные классы

data_bl.label.value_counts()
```

Уменьшились наблюдения только для самых распространенных классов _BENIGN_ и _DoS Hulk._

Теперь заменим inf значения в признаках flow_bytes/s и flow_packets/s.

```python
# заменим бесконечно большие значения в признаке flow_bytes/s максимальным значением этого признака

max_value_flow_bytes = np.nanmax(data_bl['flow_bytes/s'][data_bl['flow_bytes/s'] != np.inf ])

data_bl['flow_bytes/s'].replace([np.inf], max_value_flow_bytes, inplace= True )
```

```python
# заменим бесконечно большие значения в признаке flow_packets/s максимальным значением этого признака

max_value_flow_packets = np.nanmax(data_bl['flow_packets/s'][data_bl['flow_packets/s'] != np.inf ])

data_bl['flow_packets/s'].replace([np.inf], max_value_flow_packets, inplace= True )
```

Значения -1 у признаков `init_win_bytes_forward` и `init_win_bytes_backward` пока оставим как есть.

```python
# проверим, поменялись ли максимальные значения столбцов flow_bytes/s и flow_packets/s

data_bl.describe(include = "all")
```

```python
# проверим еще раз данные перед делением на выборки

data_bl.info()
```

### Деление на выборки, выделение признаков и целевого признака (Data splitting, features/target splitting)

Поскольку будем использовать при обучении метод кросс-валидации, то разделим датафрейм на две выборки с пропорциональным объемом имеющихся классов.

```python
# разделим данные с метками классов на обучающую (80%) и тестовую (20%)

train_bl, test_bl = train_test_split(data_bl, test_size=0.20, random_state=random_state, stratify = data_bl['label'])
```

```python
# выведем на экран размеры созданных выборок

print(f'Размер обучающей выборки: {train_bl.shape}')
print(f'Размер тестовой выборки: {test_bl.shape}')
```

```python
# создадим переменные для признаков (features) и целевого признака (target)

features_train_bl = train_bl.drop(['label'], axis=1)
target_train_bl = train_bl['label']

features_test_bl = test_bl.drop(['label'], axis=1)
target_test_bl = test_bl['label']
```

```python
# выведем на экран размеры таблиц, которые хранятся в новых переменных 

print(features_train_bl.shape)
print(target_train_bl.shape)

print(features_test_bl.shape)
print(target_test_bl.shape)
```

### Обучение моделей методом кросс-валидации (Cross-validation machine learning)

Для получения начальных метрик обучим несколько моделей: логистическую регрессию, случайный лес и бустинговую модель, например, классификатор LGBM.

#### Обучение модели логистической регрессии (LogisticRegression())

```python
# создадим пайплайн из скалера и модели логистической регрессии

lr_model = make_pipeline(StandardScaler(), LogisticRegression())
```

```python
%%time
# обучение модели логистической регрессии

cross_val_lr = cross_val_score(lr_model, features_train_bl, target_train_bl, 
                         cv = 5, 
                         n_jobs = 2, 
                         scoring = 'accuracy')


print(f'Accuracy LogisticRegression() обучающей выборки на кросс-валидации: {cross_val_lr}')
print(f'Accuracy LogisticRegression() обучающей выборки на кросс-валидации: {cross_val_lr.mean():.4%}')
```

```python
# получение предсказаний классов для целевого признака

lr_predict = cross_val_predict(lr_model, features_train_bl, target_train_bl,
                               cv=5, n_jobs = 2)

print(f'Прогнозы LogisticRegression() обучающей выборки на кросс-валидации (первые 5 прогнозов): {lr_predict[0:5]}')
```

```python
# получим Classification_report для модели логистической регрессии

score_bl_lr = classification_report(target_train_bl, lr_predict)

print(f'Classification_report модели Logistic Regression:') 
print(score_bl_lr)
```

Логистическая регрессия на не сильно обработанных данных ожидаемо не удивила.

#### Обучение модели случайного леса (RandomForestClassifier())

```python
# создадим пайплайн из скалера и модели случайного леса

rfc_model = make_pipeline(StandardScaler(), RandomForestClassifier())
```

```python
%%time
# обучение модели случайного леса

cross_val_rfc = cross_val_score(rfc_model, features_train_bl, target_train_bl, 
                         cv = 5, 
                         n_jobs = 2, 
                         scoring = 'accuracy')

print(f'Accuracy RandomForestClassifier() обучающей выборки на кросс-валидации: {cross_val_rfc}')
print(f'Accuracy RandomForestClassifier() обучающей выборки на кросс-валидации: {cross_val_rfc.mean():.4%}')
```

```python
# получение предсказаний классов для целевого признака

rfc_predict = cross_val_predict(rfc_model, features_train_bl, target_train_bl,
                               cv=5, n_jobs = 2)

print(f'Прогнозы RandomForestClassifier() обучающей выборки на кросс-валидации (первые 5 прогнозов): {rfc_predict[0:5]}')
```

```python
# получим Classification_report для модели случайного леса

score_bl_rfc = classification_report(target_train_bl, rfc_predict)

print(f'Classification_report модели RandomForestClassifier:') 
print(score_bl_rfc)
```

Случайный лес ожидаемо лучше, но всё же не идеально. На малочисленных классах метрика "проседает".

#### Обучение модели LightGBM (LGBMClassifier())

```python
# создадим пайплайн из скалера и модели LGBMClassifier

lgbm_model = make_pipeline(StandardScaler(), LGBMClassifier(class_weight='balanced'))
```

```python
%%time
# обучение модели LGBMClassifier

cross_val_lgbm = cross_val_score(lgbm_model, features_train_bl, target_train_bl, 
                         cv = 5, 
                         n_jobs = 2, 
                         scoring = 'accuracy')


print(f'Accuracy LGBMClassifier обучающей выборки на кросс-валидации: {cross_val_lgbm}')
print(f'Accuracy LGBMClassifier обучающей выборки на кросс-валидации: {cross_val_lgbm.mean():.4%}')
```

```python
# получение предсказаний классов для целевого признака

lgbm_predict = cross_val_predict(lgbm_model, features_train_bl, target_train_bl,
                               cv=5, n_jobs = 2)

print(f'Прогнозы LGBMClassifier обучающей выборки на кросс-валидации (первые 5 прогнозов): {lgbm_predict[0:5]}')
```

```python
# получим Classification_report для модели LGBMClassifier

score_bl_lgbm = classification_report(target_train_bl, lgbm_predict)

print(f'Classification_report модели LGBMClassifier:') 
print(score_bl_lgbm)
```

У LGBMClassifier при кросс-валидации видимо образовалась не очень удачная тестовая выборка, возможно в обучающую не попали те самые малочисленные классы. По итогу из-за одной плохой оценки, средняя оценка оказалась очень низкой 89.0188%, значительно ниже, чем у случайного леса.

Сведем результаты оценки бэйзлайн моделей в одну табличку.

```python
results_bl = pd.DataFrame({'model': ['LogisticRegression()', 'RandomForestClassifier()', 'LGBMClassifier()'],
'CV mean score, %': [round(cross_val_lr.mean()*100,4), round(cross_val_rfc.mean()*100,4), round(cross_val_lgbm.mean()*100,4)]})
```

```python
print(f'Результаты оценки бэйзлайн моделей на кросс-валидации (Mean accuracy, %):')
display(results_bl)
```

## Обучение лучших моделей (Best models machine learning)

Попробуем обработать данные и заново обучить модель случайного леса, а также дать второй шанс модели LGBMClassifier().

### Предобработка данных (Data preprocessing)

Теперь обработаем данные без удаления строк. Исправим inf значения на максимальные соответствующего признака. Признаки с полностью нулевыми значениями удалим. Пропуски в данных заполним с помощью _SimpleImputer_ в пайплайнах для каждой модели, используя стратегию 'mean'.

```python
# заменим бесконечно большие значения в признаке flow_bytes/s максимальным значением этого признака

max_value_flow_bytes = np.nanmax(data['flow_bytes/s'][data['flow_bytes/s'] != np.inf ])

data['flow_bytes/s'].replace([np.inf], max_value_flow_bytes, inplace= True )
```

```python
# заменим бесконечно большие значения в признаке flow_packets/s максимальным значением этого признака

max_value_flow_packets = np.nanmax(data['flow_packets/s'][data['flow_packets/s'] != np.inf ])

data['flow_packets/s'].replace([np.inf], max_value_flow_packets, inplace= True )
```

```python
# удалим столбцы (признаки) из данных, состоящие полностью из нулей (из п.1 известно, что таковых 8)

data.drop(['bwd_psh_flags', 'bwd_urg_flags', 'fwd_avg_bytes/bulk', 'fwd_avg_packets/bulk',
        'fwd_avg_bulk_rate', 'bwd_avg_bytes/bulk', 'bwd_avg_packets/bulk', 'bwd_avg_bulk_rate'], axis= 1 , inplace= True )
```

### Деление на выборки (Data splitting)

```python
# разделим данные с метками классов на обучающую (80%) и тестовую (20%)

train, test = train_test_split(data, test_size=0.20, random_state=random_state, stratify = data['label'])
```

```python
# выведем на экран размеры созданных выборок

print(f'Размер обучающей выборки: {train.shape}')
print(f'Размер тестовой выборки: {test.shape}')
```

### Генерация признаков (Feature engineering)

Создадим новый признак (feature) из статистики в разрезе меток классов, но в каждой выборке отдельно.

```python
# добавим (пока) пустой столбец со средним значением продолжительности потока в train выборке

train['flow_duration_mean'] = ''
train.head()
```

```python
# заполним столбец со средней flow_duration по классам

flow_dur_mean = train.pivot_table(values='flow_duration', index='label', aggfunc='mean')
flow_dur_mean
```

```python
for y in (flow_dur_mean.index):
    train['flow_duration_mean']= np.where(train['label'] == y, float(flow_dur_mean.loc[y]), train['flow_duration_mean'])
```

```python
train['flow_duration_mean'] = train['flow_duration_mean'].astype(float)
```

```python
train.head()
```

Теперь то же самое проделаем для тестовой выборки.

```python
# добавим (пока) пустой столбец со средним значением продолжительности потока в test выборке

test['flow_duration_mean'] = ''
test.head()
```

```python
# заполним столбец со средней flow_duration по классам

flow_dur_mean_2 = test.pivot_table(values='flow_duration', index='label', aggfunc='mean')
flow_dur_mean_2
```

```python
for y in (flow_dur_mean_2.index):
    test['flow_duration_mean']= np.where(test['label'] == y, float(flow_dur_mean_2.loc[y]), test['flow_duration_mean'])
```

```python
test['flow_duration_mean'] = test['flow_duration_mean'].astype(float)
```

```python
test.head()
```

### Выделение признаков и целевого признака (Features/target splitting)

```python
# создадим переменные для признаков (features) и целевого признака (target)

features_train = train.drop(['label'], axis=1)
target_train = train['label']

features_test = test.drop(['label'], axis=1)
target_test = test['label']
```

```python
# выведем на экран размеры таблиц, которые хранятся в новых переменных 

print(features_train.shape)
print(target_train.shape)

print(features_test.shape)
print(target_test.shape)
```

### Генерация малочисленных классов (Upsampling scarce labels target)

Для увеличения малочисленных классов в целевом признаке (Web Attack � XSS, Infiltration, Web Attack � Sql Injection, Heartbleed) скопируем имеющиеся строки для лучшего обучения модели.

```python
# создадим функцию для копирования малочисленных классов и перемешивания, 
# затем зададим параметры для получения новых выборок,


def upsample(features, target, repeat):
    features_zeros = features[target == 'Heartbleed']
    features_ones = features[target == 'Web Attack � Sql Injection']
    features_twoes = features[target == 'Infiltration']
    target_zeros = target[target == 'Heartbleed']
    target_ones = target[target == 'Web Attack � Sql Injection']
    target_twoes = target[target == 'Infiltration']
    features_threes = features[target == 'Web Attack � XSS']
    target_threes = target[target == 'Web Attack � XSS']
    

    features_upsampled = pd.concat([features_zeros] * repeat + [features_ones] * repeat + \
                                   [features_twoes] * repeat + [features_threes] * repeat)
    target_upsampled = pd.concat([target_zeros] * repeat + [target_ones] * repeat + \
                                 [target_twoes] * repeat + [target_threes] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

```python
# получим дополнительную выборку с малочисленными классами

features_upsampled, target_upsampled = upsample(features_train, target_train, 100)
```

```python
# посмотрим размер дополнительной выборки

print(features_upsampled.shape)
print(target_upsampled.shape)
```

```python
# объединим обучающую выборку с дополнительной

features_train_up = pd.concat([features_train, features_upsampled])
target_train_up = pd.concat([target_train, target_upsampled])
```

```python
# посмотрим размер получившейся обучающей выборки

print(features_train_up.shape)
print(target_train_up.shape)
```

### Обучение лучших моделей (Machine learning)

```python
%%time
# обучение модели случайного леса

cross_val_rfc = cross_val_score(rfc_model, features_train_up, target_train_up, 
                         cv = 5, 
                         n_jobs = 2, 
                         scoring = 'accuracy')


print(f'Accuracy RandomForestClassifier() обучающей выборки на кросс-валидации: {cross_val_rfc}')
print(f'Accuracy RandomForestClassifier() обучающей выборки на кросс-валидации: {cross_val_rfc.mean():.4%}')
```

```python
# получение предсказаний классов для целевого признака

rfc_predict = cross_val_predict(rfc_model, features_train_up, target_train_up,
                               cv=5, n_jobs = 2)

print(f'Прогнозы RandomForestClassifier() обучающей выборки на кросс-валидации (первые 5 прогнозов): {rfc_predict[0:5]}')
```

```python
# получим Classification_report для модели случайного леса

score_rfc = classification_report(target_train_up, rfc_predict)

print(f'Classification_report модели RandomForestClassifier:') 
print(score_rfc)
```

Метрика улучшилась с 99.7408% до 99.9888%.

#### Обучение модели LGBMClassifier() с подбором гиперпараметров с OptunaSearchCV

Сделаем подбор параметров с помощью библиотеки Optuna. OptunaSearchCV намного быстрее GridSearchCV из библиотеки sklearn.

```python
# создадим пайплайн из скалера и модели LGBMClassifier

lgbm_model = make_pipeline(StandardScaler(), LGBMClassifier(class_weight='balanced'))
```

```python
# составим список параметров для OptunaSearchCV

parameters_lgbm = {'lgbmclassifier__learning_rate': optuna.distributions.FloatDistribution(1e-2, 5e-2, step=0.02),
              'lgbmclassifier__max_depth': optuna.distributions.IntDistribution(3, 7, step=2),
              'lgbmclassifier__num_leaves': optuna.distributions.IntDistribution(10, 80, step=10),
              'lgbmclassifier__n_estimators': optuna.distributions.IntDistribution(100, 150, step=50)
             }
```

```python
%%time
# подберем параметры и обучим модель

optuna_lgbm = OptunaSearchCV(lgbm_model, param_distributions=parameters_lgbm, 
                             scoring='accuracy', cv=3, n_jobs = 2, 
                             random_state=random_state, refit=True)


optuna_lgbm.fit(features_train_up, target_train_up)
```

```python
optuna_lgbm_best_score = optuna_lgbm.best_score_
optuna_lgbm_best_params = optuna_lgbm.best_params_
optuna_lgbm_best_estimator = optuna_lgbm.best_estimator_

# лучшее значение accuracy на кросс-валидации
print(f'Accuracy LGBMClassifier() обучающей выборки: {optuna_lgbm_best_score:.4%}')
print()

# лучшие гиперпараметры
print(f'best_params: {optuna_lgbm_best_params}')
print()

# лучшая модель
print(f'best_estimator: {optuna_lgbm_best_estimator}')
print()
```

Значение accuracy не изменилось по сравнению с моделью LGBMClassifier() без подбора параметров и осталось = 99.9998%.

```python
# проверка модели на тестовой выборке

pred_test_lgbm = optuna_lgbm.predict(features_test)

print(f'Accuracy LGBMClassifier() тестовой выборки: {accuracy_score(target_test, pred_test_lgbm):.4%}')
```

```python
# получим Classification_report для модели LGBMClassifier & Optuna на тестовой выборке

score_lgbm_test = classification_report(target_test, pred_test_lgbm)
print(f'Classification_report модели LGBMClassifier & Optuna для тестовой выборки:') 
print(score_lgbm_test)
```

На тестовой выборке модель выдала более низкую точность = 99.8795%, чем на обучающей = 99.9998%.
