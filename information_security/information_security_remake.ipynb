## Обучение моделей (Machine learning)

### Предобработка данных (Data preprocessing)

Обнаруженные аномальные отрицательные значения в признаках заменим на медианные значения тех же признаков.

```python
# составим список признаков с отрицательными значениями

anomalies = ['flow_duration', 'flow_bytes/s', 'flow_packets/s', 'flow_iat_mean',
             'flow_iat_max', 'flow_iat_min', 'fwd_iat_min', 'fwd_header_length',
            'bwd_header_length', 'fwd_header_length.1', 'init_win_bytes_forward', 
             'init_win_bytes_backward', 'min_seg_size_forward']
```

```python
# заменим отрицательные значения на медианные

for anomaly in anomalies:
    data.loc[data[anomaly] < 0, anomaly] = data[anomaly].median()
```

Опытным путем было установлено, что при удалении взаимосвязанных признаков (около 40), метрики на всех моделях обрушивались на 0,1-0,2%, в т.ч.и у лог.регрессии. Тестовый набор выделялся из основного в размере **20%** от всех данных со стратификацией классов целевого признака. (Аналогично после сэмплирования выборки.)
    
Лучше всего на бэйзлайне себя показал алгоритм случайного леса. На малочисленных классах - примерно раза в 2 лучше логистической регрессии. Очень хорошая точность **precision** **=1** на `Heartbleed`, **=0.94** на `Infiltration`, **=0.96** на `Bot`, но `Web Attack � XSS`, `Web Attack � Brute Force` и `Web Attack � Sql Injection` даются плохо (**0,45-0,86**).
    
После улучшенной обработки данных и сэмплирования **LogisticRegression(), RandomForestClassifier(), LGBMClassifier()** были обучены заново. Хотя все модели улучшили свою точность, но наилучшей снова оказалась модель случайного леса: её оценки точности и полноты по каждому классу значительно выше, чем у логистической регрессии и LGBM-классификатора. Её **weighted avg F1** по сравнению со своим же бэйзлайном не изменился (=1). Распознавание малочисленных классов улучшилось:
 - `Web Attack � Brute Force` **с 0,71 до 0,90**
 - `Web Attack � Sql Injection` **с 0,86 до 0,99**
 - `Web Attack � XSS` с **0,45 до 0,75**
 - `Infiltration` **с 0,94 до 1**
 - `Bot` **с 0,96 до 0,98**

При этом точность определения других классов не упала.
    
В результате многочисленных экспериментов пришлось отказаться от удаления взаимно коррелирующих признаков **(мультиколлинеарности)**, т.к.они при этом коррелировали с целевой переменной и при их удалении сильно обрушивались метрики (в т.ч.и у лог.регрессии).

Также в этом исследовании были полностью удалены **явные дубликаты**, а **сэмплирование данных** производилось при помощи метода **SMOTETomek** (не копирование, а генерирование похожих наблюдений в условиях избыточной и недостаточной выборки) в отличие от первого исследования, где дубликаты не удалялись, а данные малочисленных классов дополнительно множились простым копированием. 
    
Тем не менее добиться метрик **precision, recall, f1_score = 1** по всем классам не удалось, в отличие от первого исследования, где модели **RandomForestClassifier() и LGBMClassifier()** добились такого результата.
