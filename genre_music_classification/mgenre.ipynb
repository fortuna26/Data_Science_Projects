### Признаки

В датасете имеются следующие признаки в виде непрерывных величин: acousticness, danceability, duration_ms, energy, instrumentalness, liveness, loudness, speechiness, tempo, valence. Причем все из них, кроме duration_ms, loudness, tempo, имеют значения от 0 до 1 и не имеют выбросов за эти пределы. Признак instrumentalness/инструментальность имеет много нулевых значений. Пока непонятно выбросы это или норма.

Признак duration_ms/длительность имеет большой разброс и присутствуют аномальные значения -1. Причем как в обучающей, так в тестовой выборке.

Признак loudness/громкость имеет как положительные, так и отрицательные значения, но отрицательные в большей степени.

У признака tempo/темп разброс значений от 34 до 220. Выбросы надо смотреть отдельно.

Также в датасете есть категориальные признаки: key - 12 уникальных значений и mode - 2 уникальных значения.

Признак obtained_date/дата загрузки в сервис имеет всего 4 уникальных значения и не имеет значения для классификации жанров композиций.

Признаки instance_id и track_name - уникальный номер трека и его название, соответственно уникальны (по количеству наблюдений), их много и они вряд ли могут помочь модели классифицировать жанры.

```python
# сгруппируем в обучающей выборке треки по жанрам и посмотрим на их статистики инструментальности.

genre_train.groupby('music_genre')['instrumentalness'].agg({'count', 'median', 'mean'}) \
           .sort_values(('mean'), ascending=False) \
           .reset_index() \
           .rename(columns={'count': 'всего треков', 'median': 'медиана инструментальности звучания',
                            'mean': 'средняя инструментальность звучания', 'music_genre': 'музыкальный жанр'})
```

В обучающей выборке key перекликается с mode (-0.04), а в тестовой у key есть отдаленная корреляция с acousticness, speechiness, и также с mode. Используем корреляцию между key и mode для обеих выборок.

Попробуем найти взаимные закономерности между key и mode. Т.к.это категориальные переменные, найдем их моды в разрезе друг друга.

```python
# категоризируем непрерывную величину признака tempo

def categorize_tempo(tempo):
    try:
        if 0 <= tempo <= 30.999:
            return '1'
        elif 31.000 <= tempo <= 50.999:
            return '2'
        elif 51.000 <= tempo <= 60.999:
            return '3'
        elif 61.000 <= tempo <= 70.999:
            return '4'
        elif 71.000 <= tempo <= 80.999:
            return '5'
        elif 81.000 <= tempo <= 90.999:
            return '6'
        elif 91.000 <= tempo <= 100.999:
            return '7'
        elif 101.000 <= tempo <= 110.999:
            return '8'
        elif 111.000 <= tempo <= 120.999:
            return '9'
        elif 121.000 <= tempo <= 130.999:
            return '10'
        elif 131.000 <= tempo <= 140.999:
            return '11'
        elif 141.000 <= tempo <= 150.999:
            return '12'
        elif 151.000 <= tempo <= 160.999:
            return '13'
        elif 161.000 <= tempo <= 170.999:
            return '14'
        elif 171.000 <= tempo <= 200.999:
            return '15'
        elif tempo >= 201.000:
            return '16'
    except:
        pass
```

Попробуем также определить интервалы энергичности для заполнения таблицы пропущенными значениями tempo_category, а потом уже на основании этих значений, заполнить какой-нибудь внутрикатегорийной статистикой признак tempo.
Возможно, это будет лучше, чем просто заполнить нулями. Хотя в целом, пропусков в этих признаках (tempo и tempo_category) и не так много, чтобы кардинально повлиять на качество обучения модели. Но будем проверять.

```python
# категоризируем tempo_category непрерывную величину признака energy, чтобы потом заменить ею пропуски в tempo_category

def categorize_energy(energy):
    try:
        if 0 <= energy <= 0.399999:
            return '4'                                        # больше всего треков из 3-х категорий (4, 3, 2)
        elif 0.400000 <= energy <= 0.589999:
            return '6'                                        # больше всего треков из 2-х категорий (5, 6)
        elif 0.590000 <= energy <= 0.689999:
            return '7'                                        # больше всего треков из 5-и категорий (7, 8, 9, 11, 16)
        elif 0.690000 <= energy <= 0.729999:
            return '10'                                       # больше всего треков из 2-х категорий (10, 12)
        elif 0.730000 <= energy <= 1.0000:
            return '15'                                       # больше всего треков из 3-х категорий (15, 13, 14)  
    except:
        pass
```
